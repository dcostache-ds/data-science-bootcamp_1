3. Best Time to Buy and Sell Stock — Sliding Window Dai un array cu prețurile unei acțiuni pe fiecare zi. Returnează profitul maxim posibil. Trebuie să cumperi înainte să vinzi.
Input: prices = [7, 1, 5, 3, 6, 4]
Output: 5
Explicatie: Cumperi la 1, vinzi la 6 → profit = 5

Input: prices = [7, 6, 4, 3, 1]
Output: 0
Explicatie: Preturile scad mereu, nu faci profit


```python
prices = [7, 1, 5, 3, 6, 4]
min_price=float('inf')
max_profit=0
for price in prices:
    min_price=min(min_price,price)
    profit=price-min_price
    max_profit=max(max_profit,profit)
print(max_profit)
```

```python
nums = [-1, 0, 1, 2, -1, -4]
nums.sort()
res = []
print(nums)
for i,val in enumerate(nums):
    if i > 0 and nums[i-1]==val:
        continue 
    l,r=i+1,len(nums)-1
    while l<r:
        suma=nums[l]+nums[r]+val
        if suma>0:
            r-=1
        elif suma<0:
            l+=1
        else:
            res.extend([nums[l],nums[r],val])
            l+=1
            while nums[l] == nums[l-1] and l<r:
                l+=1
        
print(res)

    
```

9. Merge Intervals — Sorting Dată o listă de intervale, unește toate intervalele care se suprapun.
Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]
Explicatie: [1,3] si [2,6] se suprapun → [1,6]

Input: intervals = [[1,4],[4,5]]
Output: [[1,5]]
Explicatie: Se ating la 4 → se unesc

Input: intervals = [[4,7],[1,4]]
Output: [[1,7]]


```python
intervals = [[1,3],[2,6],[8,10],[15,18]]
intervals.sort()
res=[intervals[0]]
for i in range(1,len(intervals)):
    last_end=res[-1][1]
    current_end=intervals[i][1]
    current_start=intervals[i][0]
    if last_end >= current_start:
        res[-1][1]=max(last_end, current_end)
    else:
        res.append(intervals[i])

print(res)

    

```

11. Reverse Linked List — Linked List Inversează o listă înlănțuită și returnează noul head.
Input:  1 → 2 → 3 → 4 → 5
Output: 5 → 4 → 3 → 2 → 1

Input:  1 → 2
Output: 2 → 1


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev=None
        current=head

        while current:
            next_node=current.next
            current.next=prev
            prev=current
            current=next_node
        print(prev)

```

13. Container With Most Water — Two Pointers Dat un array cu înălțimi de linii verticale, găsește două linii care formează containerul cu cea mai mare cantitate de apă.
Input: height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
Output: 49

Input: height = [1, 1]
Output: 1


```python
height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
arie_maxima=0
l,r=0,len(height)-1
while l<r:
    diferenta= r-l
    minimul=min(height[l],height[r])

    produs=diferenta*minimul
    arie_maxima = max(arie_maxima, produs)
    if height[l] < height[r]:
        l+=1
    else:
        r-=1
print(arie_maxima)
```
