### Step 14. How much was the revenue for the period in the dataset?

```python
chipo['revenue']=chipo['quantity']*chipo['item_price']
chipo['revenue'].sum()
venit_per_produs = chipo.groupby('item_name')['revenue'].sum()

# O sortam ca sa vedem topul
venit_per_produs = venit_per_produs.sort_values(ascending=False)

print(venit_per_produs)
total_revenue = chipo["revenue"].sum()

```

```python
# Extragi direct numele celui care are valoarea maxima
nume_produs = venit_per_produs.idxmax()

# Extragi direct valoarea maxima matematica
bani_incasati = venit_per_produs.max()

print("Cel mai banos produs este:", nume_produs)
print("Venitul generat:", bani_incasati)

# sau folosesc reset_index ca sa l fac dataframe si apoi accesez cu iloc[0,0]
tabel_venit=venit_per_produs.reset_index()
tabel_venit.iloc[0,1]
```
