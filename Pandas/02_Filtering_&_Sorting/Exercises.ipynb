### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv('chipotle.tsv',sep='\t')
chipo
chipo["item_price"] = chipo["item_price"].apply(lambda x: float(x[1:]))
```

### Step 4. How many products cost more than $10.00?

```python
#chipo['item_price']=chipo['item_price'].str.replace("$","").astype(float)
products_more_than=chipo[chipo['item_price']>10.00].value_counts()
len(products_more_than)
```

### Step 5. What is the price of each item? 
###### print a data frame with only three columns item_name choice_description and product_price

```python
price_for_each=chipo[['item_name','choice_description','item_price']].drop_duplicates()
price_for_each
#drop_duplicates() -> Arata-mi TABELUL curatat
price_unique=chipo[['item_name','choice_description','item_price']].nunique()
price_for_each
#nunique-> imi da doar numarul lor cate is
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name').reset_index()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
index_max=chipo["item_price"].idxmax()

chipo['quantity'].iloc[index_max]
```

```python
max_item_price_row=chipo[chipo['item_price']==chipo['item_price'].max()].reset_index()
max_item_price_row['quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

### Step 9. How many times did someone order more than one Canned Soda?

```python
soda_ordered=chipo[chipo['item_name'] == 'Canned Soda'].value_counts().reset_index()
nr_soda=soda_ordered[soda_ordered['quantity']>1].value_counts()
len(nr_soda)
```
