### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv('chipotle.tsv',sep='\t')
chipo.head()
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
items=chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)

items.plot(kind='bar')
plt.title('TOp 5 produse cumparate')

plt.xlabel('Nume Produs')
plt.ylabel('NR Produse cumparate')
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
comenzi = chipo.groupby('order_id').sum()

plt.scatter(x = comenzi['item_price'], y = comenzi['quantity'])

plt.xlabel('Pretul Comenzii')
plt.ylabel('Numarul de produse pe comanda')
plt.title('Numar produse vs. Pret comanda')

plt.show()
```
